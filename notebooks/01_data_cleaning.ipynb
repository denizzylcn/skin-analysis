```python
# 3) Veri klasörlerini Drive'dan bağla
import shutil, os
from pathlib import Path

# Drive'daki datasetleri projeye symlink yap
datasets = {
    'skin_type' : f'{BASE}/Oily-Dry-Skin-Types',
    'acne'      : f'{BASE}/skin',
    'lesion'    : f'{BASE}/deri lezyonu',
}
for name, src in datasets.items():
    dst = f'data/raw/{name}'
    if not os.path.exists(dst):
        os.symlink(src, dst)
        print(f'Baglandi: {dst}')
    else:
        print(f'Zaten var: {dst}')
```
